```python
import numpy
values = [4, 5, 20, 22, 22, 24, 24, 26, 27, 29, 29]
x = numpy.percentile(values,25)
print(x)
```

```python
import numpy as np
values = np.array([4, 5, 20, 22, 22, 24, 24, 26, 27, 29, 29])
p25 = np.percentile(values,25,method='lower')
print(p25)
```

```python
p75 = np.percentile(values,75,method='lower')
print(p75)
```

```python
IQR = p75-p25
print(IQR)
```
